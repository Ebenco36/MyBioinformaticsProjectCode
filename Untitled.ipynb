```python
from sklearn.model_selection import train_test_split
train_data, test_data, train_lbl, test_lbl = train_test_split( df_data.loc[:,df_data.columns != 'labels'].values, 
                                                              final_df['labels'], 
                                                              test_size=1/7.0, 
                                                              random_state=42, shuffle=True)
```

```python
scaler = StandardScaler()
# Fit on training set only.
scaler.fit(train_data)
# Apply transform to both the training set and the test set.
train_data_ = scaler.transform(train_data)
test_data_ = scaler.transform(test_data)
```

```python
train_data.shape
```

```python
df_data.shape
```

```python
from sklearn.decomposition import PCA
# Make an instance of the Model
pca = PCA(.60)
pca.fit(train_data_)
pca.n_components_
```

```python
train_data_transform = pca.transform(train_data_)
test_data_transform = pca.transform(test_data_)
```

```python
train_data_transform
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
# logisticRegr = LogisticRegression(solver = 'lbfgs')
pipe = make_pipeline(StandardScaler(), LogisticRegression())
pipe_svm = make_pipeline(StandardScaler(), SVC())

pipe_rdf = make_pipeline(StandardScaler(), RandomForestClassifier())
pipe_lda = make_pipeline(StandardScaler(), LinearDiscriminantAnalysis())
```

```python
# logisticRegr.fit(train_data_transform, train_lbl)
pipe.fit(train_data_transform, train_lbl)  # apply scaling on training data
pipe_svm.fit(train_data_transform, train_lbl)  # apply scaling on training data
pipe_rdf.fit(train_data_transform, train_lbl)  # apply scaling on training data
pipe_lda.fit(train_data_transform, train_lbl)  # apply scaling on training data

# logisticRegr.predict(test_data_transform)
score_log = pipe.score(test_data_transform, test_lbl)
predict_log = pipe.predict(test_data_transform)

score_svm = pipe_svm.score(test_data_transform, test_lbl)
predict_svm = pipe_svm.predict(test_data_transform)

score_rdf = pipe_rdf.score(test_data_transform, test_lbl)
predict_rdf = pipe_rdf.predict(test_data_transform)

score_lda = pipe_lda.score(test_data_transform, test_lbl)
predict_lda = pipe_lda.predict(test_data_transform)

print('Log Score: {}, Log Predict: {}, SVM Score: {}, SVM Predict: {}, Random Score: {}, Random Predict: {}, LDA Score: {}, LDA Predict: {}'.format(
        score_log, predict_log, score_svm, predict_svm, score_rdf, predict_rdf, score_lda, predict_lda))
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, precision_score, recall_score
from sklearn import metrics

confusion_matrix(test_lbl, predict_svm)
```

```python
def get_metrics(y_test, y_predicted):
    accuracy = accuracy_score(y_test, y_predicted)
    precision = precision_score(y_test, y_predicted, average='weighted')
    recall = recall_score(y_test, y_predicted, average='weighted')
    f1 = f1_score(y_test, y_predicted, average='weighted')
    return accuracy, precision, recall, f1
accuracy, precision, recall, f1 = get_metrics(test_lbl, predict_log)
print("accuracy = %.3f \nprecision = %.3f \nrecall = %.3f \nf1 = %.3f" % (accuracy, precision, recall, f1))
```
